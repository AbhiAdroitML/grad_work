**Reading** **Merged** **Output** **File**

```python
# read the merged output from previous step
mergedData = pd.read_pickle(os.path.join(path,'2_mergedData.pkl'))
```

```python
mergedData['date'] = pd.to_datetime(mergedData['reviewTime']) # review date from UNIX format
```

```python
mergedData['lenReview']= mergedData['reviewText'].str.split().str.len() # finding number of words in each review
```

```python
len(mergedData)
```

**Sampling** **Data**

```python
# 1. Filter on length of review (minimum 10 words)
mergedDataSubset = mergedData.query("lenReview > 10")
```

```python
# 2. Filter on time duration (year greater than 2015)
TopBrands = mergedDataSubset[mergedDataSubset.date.dt.year>=2015]
```

```python
# 3. Drop all duplicate reviews
TopBrands = TopBrands.drop_duplicates(ignore_index=True, keep='first', subset='reviewText')
```

**Finding Top Brands in Filtered Data**

```python
TopBrands.groupby(TopBrands.date.dt.year)['asin'].count().sort_values(ascending=False)
```

```python
len(TopBrands)
```

```python
brands = (TopBrands.groupby('brand')['asin'].count().sort_values(ascending=False).head(5)).index.tolist()
```

```python
Top5Brands = TopBrands[TopBrands['brand'].isin(brands)]
print( len(Top5Brands))
```

**Restricting Sample Size to 100K**

```python
Top5BrandSample = Top5Brands.sample(n=100000, random_state=1)
print(len(Top5BrandSample))
```
